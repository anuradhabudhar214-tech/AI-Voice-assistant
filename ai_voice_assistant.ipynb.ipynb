```
!pip install gtts

from gtts import gTTS
from IPython.display import Audio, display

# Simple knowledge base (RAG)
knowledge_base = {
    "dubai mall": "Dubai Mall is one of the largest shopping malls in Dubai.",
    "burj khalifa": "Burj Khalifa is the tallest building in the world.",
    "python": "Python is commonly used in AI and data science.",
    "machine learning": "Machine learning allows systems to learn from data.",
    "ai assistant": "An AI assistant helps users by answering questions."
}

def get_response(user_text):
    text = user_text.lower()

    # Check knowledge base first (prevents hallucination)
    for key in knowledge_base:
        if key in text:
            return knowledge_base[key]

    # Simple conversation logic
    if "hello" in text:
        return "Hello! How can I help you today?"
    elif "help" in text:
        return "I can provide information about our services."
    elif "bye" in text:
        return "Thank you for your time. Have a good day!"
    else:
        return "Sorry, I don’t have information on that."

def speak(text):
    try:
        tts = gTTS(text=text, lang="en")
        tts.save("audio.mp3")
        display(Audio("audio.mp3", autoplay=True))
    except:
        print("(Audio unavailable, text response shown)")

print("AI Voice Assistant started")
print("Type your message below. Type 'bye' to exit.")

# Simulated real-time STT using text input
while True:
    user_input = input("User:")
    response = get_response(user_input)
    print("AI:", response)
    speak(response)

    if "bye" in user_input.lower():
        break
```
